## Read and Resize Images

```python
# import libraries
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# reading images
img1 = cv2.imread('rock.jpg')
img2 = cv2.imread('beach.jpg')
# make the images in the same size
dim = (500, 500)
img1r = cv2.resize(img1, dim)
img2r = cv2.resize(img2, dim)
print(img1r, img2r)
```

Dengan melakukan *input* di atas, kedua citra sudah dimuat dengan **ukuran atau dimensi matriksnya 500 Ã— 500** dan siap untuk dilakukan pemrosesan lebih lanjut.

## Display Images

```python
# show image using matplotlib.pyplot
# create figure
fig = plt.figure(figsize=(10, 10))
  
# setting values to rows and column variables
rows = 2
columns = 2
    
# Adds a subplot at the 1st position
fig.add_subplot(rows, columns, 1)
  
# showing image
plt.imshow(img1r)
plt.axis('off')
plt.title("Image 1 by matplotlib")
  
# Adds a subplot at the 2nd position
fig.add_subplot(rows, columns, 2)
  
# showing image
plt.imshow(img2r)
plt.axis('off')
plt.title("Image 2 by matplotlib")
```

Ada perbedaan dalam pengurutan piksel di *OpenCV* dan *Matplotlib*. *OpenCV* mengikuti urutan BGR, sementara ***matplotlib* mengikuti urutan RGB**. Oleh karena itu, ketika menampilkan citra yang dimuat di *OpenCV* menggunakan fungsi *matplotlib*, harus dilakukan **pengubahan menjadi mode RGB**.

```python
# convert to RGB
rgb1 = cv2.cvtColor(img1r, cv2.COLOR_BGR2RGB)
rgb2 = cv2.cvtColor(img2r, cv2.COLOR_BGR2RGB)
```

Dengan menjalankan *syntax* di atas, **citra telah berubah menjadi mode RGB** sesuai dengan urutan di *matplotlib* dan citra telah sesuai seperti citra aslinya sebagai berikut.

```python
# show image again using matplotlib.pyplot
# create figure
fig = plt.figure(figsize=(10, 10))
  
# setting values to rows and column variables
rows = 2
columns = 2
    
# Adds a subplot at the 1st position
fig.add_subplot(rows, columns, 1)
  
# showing image
plt.imshow(rgb1)
plt.axis('off')
plt.title("Image 1 by OpenCV")
  
# Adds a subplot at the 2nd position
fig.add_subplot(rows, columns, 2)
  
# showing image
plt.imshow(rgb2)
plt.axis('off')
plt.title("Image 2 by OpenCV")
```

## Describe Images
Dalam melakukan pemrosesan citra (*image processing*), seringkali perlu mengetahui ukuran citra yang dimuat atau ditransformasikan melalui berbagai tahapan.
Saat menggunakan *OpenCV Python*, citra disimpan dalam *NumPy N-dimensional array*. Untuk mendapatkan bentuk atau ukuran citra, digunakan *ndarray.shape* untuk mendapatkan dimensi citra. Kemudian, indeks pada variabel dimensi dapat digunakan untuk mendapatkan lebar (*width*), tinggi (*height*), dan jumlah saluran (*channels*) untuk setiap piksel.
* **Tinggi (*height*)** merepresentasikan jumlah baris piksel dalam citra atau jumlah piksel di setiap kolom citra.
* **Lebar (*width*)** merepresentasikan jumlah kolom piksel dalam citra atau jumlah piksel di setiap baris citra.
* **Jumlah saluran (*channels*)** merepresentasikan jumlah komponen yang digunakan untuk mewakili setiap piksel.

```python
# describing images
# get dimensions of image
dim1 = img1r.shape
dim2 = img1r.shape

# height, width, number of channels in images
height1 = img1r.shape[0]
width1 = img1r.shape[1]
channels1 = img1r.shape[2]

print('Dimension of Image 1   : ',dim1)
print('Height of Image 1      : ',height1)
print('Width of Image 1       : ',width1)
print('Number of Channels     : ',channels1)

height2 = img2r.shape[0]
width2 = img2r.shape[1]
channels2 = img2r.shape[2]

print('Dimension of Image 2   : ',dim2)
print('Height of Image 2      : ',height2)
print('Width of Image 2       : ',width2)
print('Number of Channels     : ',channels2)
```

Metode *shape* objek *array numpy* menemukan jumlah baris (tinggi) dan jumlah kolom (lebar) kedua citranya sebanyak 500 dengan jumlah *channel* sebanyak 3, yaitu R (*Red*), G (*Green*), dan B (*Blue*).

## Image Addition
Operasi penambahan pada citra dapat dilakukan dengan menambahkan dua citra atau menambahkan nilai konstan ke citra. Operasi penambahan pada citra menggunakan *OpenCV* dilakukan dengan memuat *file* citra dan memasukkan *array NumPy N-d* yang didapat setelah memuat citra ke fungsi *cv2.add()* sebagai argumen. Ini adalah operasi tersaturasi yang berarti jika nilai piksel yang dihasilkan lebih besar dari 255 setelah penambahan nilai piksel dari citra input, maka hasil disaturasi menjadi 255 sehingga **nilai piksel apa pun tidak dapat melebihi 255**. Ini disebut *saturation.* Misalnya: 250 + 10 = 260 menjadi 255.

```python
# Addition of Images
add = cv2.add(img1r, img2r)
print(add)
addrgb = cv2.cvtColor(add, cv2.COLOR_BGR2RGB)
plt.axis('off')
plt.imshow(addrgb)
plt.show()
```

Diperoleh penambahan *image img1r* dan *img2r* menghasilkan citra di atas dengan nilai-nilai matriksnya merupakan hasil penjumlahan tiap-tiap elemen kedua matriks yang bersesuaian.

## Image Subtraction
Operasi pengurangan pada citra adalah pengurangan piksel yang mengambil dua citra sebagai *input* dan menghasilkan citra ketiga sebagai *output* yang nilai pikselnya sama dengan citra pertama dikurangi nilai piksel yang sesuai dari citra kedua. Satu citra juga dapat digunakan sebagai input dan semua nilai pikselnya dengan nilai konstan. Beberapa versi operator akan menampilkan perbedaan mutlak antara nilai piksel, daripada keluaran langsung yang ditandai.

Implementasi pengurangan citra bervariasi seperti apa yang operatornya lakukan jika nilai piksel keluaran negatif. Jika format citra mendukung nilai piksel negatif, dalam hal ini, nilai negatifnya dibiarkan saja. 
**Jika format citra tidak mendukung nilai piksel negatif, seringkali piksel tersebut dijadikan nol (biasanya hitam).**

Jika pengurangan citra menghitung selisih mutlak dari dua citra masukan yang menggunakan jenis nilai piksel yang sama, maka **nilai piksel keluaran tidak boleh berada di luar rentang yang ditentukan** yang mungkin diwakili oleh jenis piksel citra masukan sehingga masalah ini tidak muncul. Itu sebabnya ada baiknya menggunakan perbedaan absolut. 

```python
# Subtraction of Images
sub = cv2.subtract(img1r, img2r)
print(sub)
subrgb = cv2.cvtColor(sub, cv2.COLOR_BGR2RGB)
plt.axis('off')
plt.imshow(subrgb)
plt.show()
```

Diperoleh pengurangan *image img1r* dan *img2r* menghasilkan citra di atas dengan nilai-nilai matriksnya merupakan hasil pengurangan tiap-tiap elemen kedua matriks yang bersesuaian.

## Image Multiplication
Seperti operasi aritmatika lainnya pada citra, perkalian citra juga dapat diimplementasikan. Bentuk pertama dari perkalian citra mengambil dua citra masukan dan menghasilkan sebuah citra keluaran dimana nilai pikselnya merupakan perkalian dari nilai piksel yang bersesuaian dari citra masukan, sedangkan bentuk kedua mengambil citra masukan tunggal dan menghasilkan keluaran di mana setiap nilai piksel adalah produk dari nilai piksel yang sesuai dari citra masukan dan konstanta yang ditentukan (faktor penskalaan). Bentuk perkalian citra kedua ini lebih banyak digunakan dan umumnya disebut *scaling.*
Ada beberapa kegunaan penskalaan citra (*scaling*) tetapi secara umum **faktor penskalaan bernilai lebih besar dari satu akan mencerahkan citra dan faktor penskalaan yang kurang dari satu akan menggelapkan citra** dengan catatan bahwa nilai konstan seringkali berupa angka *floating-point*, tergantung pada intensitas citra yang dapat ditingkatkan atau diturunkan. Nilai bisa menjadi negatif jika format citra mendukung hal ini. Jika nilai keluaran dihitung lebih besar dari nilai piksel maksimum yang dibatasi, maka nilai tersebut dipotong pada nilai piksel maksimum yang dibatasi.

```python
# Multiplication of Images 
mult = cv2.multiply(img1r, img2r)
print(mult)
multrgb = cv2.cvtColor(mult, cv2.COLOR_BGR2RGB)
plt.axis('off')
plt.imshow(multrgb)
plt.show()
```

Diperoleh perkalian *image img1r* dan *img2r* menghasilkan citra di atas dengan sebagian besar nilai elemen matriksnya mencapai batas maksimum, yaitu nilai 255 yang berwarna putih.

Selain mengalikan antara citra 1 dan 2, perkalian juga dapat dilakukan dengan mengalikan citra dengan skalar seperti yang dilakukan sebagai berikut.

```python
# Multiplication by Scalar
ms1 = cv2.multiply(img1r, 1.5)
print(ms1)
ms1rgb = cv2.cvtColor(ms1, cv2.COLOR_BGR2RGB)
plt.axis('off')
plt.imshow(ms1rgb)
plt.show()
ms2 = cv2.multiply(img2r, 1.5)
print(ms2)
ms2rgb = cv2.cvtColor(ms2, cv2.COLOR_BGR2RGB)
plt.axis('off')
plt.imshow(ms2rgb)
plt.show()
```

Diperoleh perkalian *image img1r* dan *img2r* dengan nilai 1,5 menghasilkan citra di atas dengan nilai-nilai matriksnya merupakan hasil perkalian tiap-tiap elemen matriks citra dengan nilai 1,5. 

## Image Division

```python
# Division of Images
div = cv2.divide(img1r, img2r)
print(div)
divrgb = cv2.cvtColor(div, cv2.COLOR_BGR2RGB)
plt.axis('off')
plt.imshow(divrgb)
plt.show()
```

Diperoleh pembagian *image img1r* dengan *img2r* menghasilkan citra di atas dengan nilai-nilai matriksnya merupakan hasil penjumlahan tiap-tiap elemen kedua matriks yang bersesuaian.

Selain melakukan pembagian antara citra 1 dan 2, pembagian juga dapat dilakukan dengan membagi citra dengan skalar seperti yang dilakukan sebagai berikut.

```python
# Division by Scalar
div1 = cv2.divide(img1r, 2)
print(div1)
div1rgb = cv2.cvtColor(div1, cv2.COLOR_BGR2RGB)
plt.axis('off')
plt.imshow(div1rgb)
plt.show()
div2 = cv2.divide(img2r, 2)
print(div2)
div2rgb = cv2.cvtColor(div2, cv2.COLOR_BGR2RGB)
plt.axis('off')
plt.imshow(div2rgb)
plt.show()
```

Diperoleh pembagian *image img1r* dan *img2r* dengan nilai 2 menghasilkan citra di atas dengan nilai-nilai matriksnya merupakan hasil pembagian tiap-tiap elemen matriks citra dengan nilai 2.
