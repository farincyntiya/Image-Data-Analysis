# Morphological Transformations

<b>Morphological transformations</b> are some simple operations based on the image shape and are normally performed on binary images. 
There are two things needed for this operation, such as <b>the original image and kernel.</b> A kernel tells how to change the value of any given pixel by combining it with different amounts of the neighboring pixels.

## Importing Libraries

```python
# working directory
import os
os.getcwd()
# import library
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

## Importing Image and Creating Mask

```python
# import image as grayscale
img = cv2.imread('smarties.png', cv2.IMREAD_GRAYSCALE)

# provide a mask so that the image is binary
# image, the value of treshold, maximum value of threshold, type of the threshold
_, mask = cv2.threshold(img, 220, 255, cv2.THRESH_BINARY_INV)

# see how the image looks like after applying the mask
titles = ['images', 'mask']
images = [img, mask]
```

## Visualizing the Image

```python
# range two because now the array is of two elements
for i in range(2):
    plt.subplot(1, 2, i + 1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
    
plt.show()    
```

After masking, there are some black dots inside the balls in white area. Those black dots can be removed by using <b>dilation</b>.

## Defining a Kernel

```python
# define a kernel
# normally a square or some shape which then applied to the image
# kernel with size 2 by 2 square shape
kernel = np.ones((2,2), np.uint8)
```

## Dilation

```python
# dilation
dilation = cv2.dilate(mask, kernel)

# see how the image looks like after applying mask and dilation
titles = ['images', 'mask', 'dilation']
images = [img, mask, dilation]

# range three because now the array is of three elements
for i in range(3):
    plt.subplot(1, 3, i + 1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
    
plt.show() 
```

After applying dilation, some black dots are reduced (by size or by deletion). These black dots can be removed completely by doing <b>iterations on dilation.</b>

```python
# dilation with iterations
dilation = cv2.dilate(mask, kernel, iterations = 2)

# see how the image looks like after applying mask and dilation
titles = ['images', 'mask', 'dilation']
images = [img, mask, dilation]

# range three because now the array is of three elements
for i in range(3):
    plt.subplot(1, 3, i + 1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
    
plt.show() 
```

After applying <b>dilation with 2 iterations</b>, some black dots are now gradually gone but there are still some small black dots.

```python
# the size of the kernel can be increased 
# the bigger, the better the result will be, but there also be a problem
kernel = np.ones((5,5), np.uint8)

# dilation
dilation = cv2.dilate(mask, kernel, iterations = 2)

# see how the image looks like after applying mask and dilation
titles = ['images', 'mask', 'dilation']
images = [img, mask, dilation]

# range three because now the array is of three elements
for i in range(3):
    plt.subplot(1, 3, i + 1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
    
plt.show() 
```

With bigger size of kernel, now all the black dots from the image is gone. But, it is also observed that the size of the white area also increased after applying the dilation on the masked image.

There all balls that seems merged after dilation, because the size of the kernel is big and the pixel element is one if at least one pixel <b>under the kernel</b> is one, that's why the shape of those balls are increasing.

## Erosion

```python
# erosion
# by default, the iteration is also one
erosion = cv2.erode(mask, kernel, iterations = 1)

# see how the image looks like after applying mask and dilation
titles = ['images', 'mask', 'dilation', 'erosion']
images = [img, mask, dilation, erosion]

# range four because now the array is of four elements
# displaying with 1 by 4
for i in range(4):
    plt.subplot(1, 4, i + 1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
    
plt.show() 

# displaying with 2 by 2
for i in range(4):
    plt.subplot(2, 2, i + 1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
    
plt.show() 
```

With dilation, all the spots in the balls which are black are gone, but the size is increased.

With erosion, the size of the balls eroded so the basic idea of erosion is just like soil erosion, it erodes only the boundary of the foreground object. So when erosion is applied, the kernel which have been defined, slides through all the image and <b>a pixel in the original image either 1 or 0 will be considered as 1 only if all the pixels under the kernel is 1, otherwise it is eroded and this means this value will be set to 0 and will be a black area.</b>

```python
# dilation with iterations
erosion = cv2.erode(mask, kernel, iterations = 2)

# see how the image looks like after applying mask and dilation
titles = ['images', 'mask', 'dilation', 'erosion']
images = [img, mask, dilation, erosion]

# range four because now the array is of four elements
# displaying with 2 by 2
for i in range(4):
    plt.subplot(2, 2, i + 1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
    
plt.show() 
```

Now, the balls are eroded more.

```python
# dilation with iterations
erosion = cv2.erode(mask, kernel, iterations = 5)

# see how the image looks like after applying mask and dilation
titles = ['images', 'mask', 'dilation', 'erosion']
images = [img, mask, dilation, erosion]

# range four because now the array is of four elements
# displaying with 2 by 2
for i in range(4):
    plt.subplot(2, 2, i + 1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
    
plt.show() 
```

Now, the balls are really small because multiply number of times of erosion have been applied.

```python
# the size of the kernel can be decreased
# the bigger, the better the result will be, but there also be a problem
kernel = np.ones((2,2), np.uint8)

# dilation with iterations
erosion = cv2.erode(mask, kernel, iterations = 5)

# see how the image looks like after applying mask and dilation
titles = ['images', 'mask', 'dilation', 'erosion']
images = [img, mask, dilation, erosion]

# range four because now the array is of four elements
# displaying with 2 by 2
for i in range(4):
    plt.subplot(2, 2, i + 1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
    
plt.show() 
```

Now, the result is better because all the spots from the balls are gone (dilation) and the balls are not so much eroded (erosion).

## Opening

```python
# opening
# provide the sources: mask, the type of morphological operation, kernel
opening = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)

# see how the image looks like after applying mask and dilation
titles = ['images', 'mask', 'dilation', 'erosion', 'opening']
images = [img, mask, dilation, erosion, opening]

# range five because now the array is of five elements
# displaying with 2 by 3
for i in range(5):
    plt.subplot(2, 3, i + 1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
    
plt.show() 
```

Opening is another name of <b>erosion followed by dilation.</b> First of all, erosion is performed on the image and then the dilation is performed on the image. Still it is seen some spots which can be gone by <b>increasing the kernel size.</b>

```python
# the size of the kernel can be increased 
# the bigger, the better the result will be, but there also be a problem
kernel = np.ones((5,5), np.uint8)

# opening
# provide the sources: mask, the type of morphological operation, kernel
opening = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)

# see how the image looks like after applying mask and dilation
titles = ['images', 'mask', 'dilation', 'erosion', 'opening']
images = [img, mask, dilation, erosion, opening]

# range five because now the array is of five elements
# displaying with 2 by 3
for i in range(5):
    plt.subplot(2, 3, i + 1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
    
plt.show() 
```

Now, the image somehow looks better. So, <b>opening is erosion followed by dilation.</b>

## Closing

Closing is the opposite of opening. In the closing morphological transformation, <b>dilation is performed first on the image and then it is followed by the erosion.<b>

```python
# closing
# provide the sources: mask, the type of morphological operation, kernel
closing = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)

# see how the image looks like after applying mask and dilation
titles = ['images', 'mask', 'dilation', 'erosion', 'opening', 'closing']
images = [img, mask, dilation, erosion, opening, closing]

# range six because now the array is of six elements
# displaying with 3 by 3
for i in range(6):
    plt.subplot(2, 3, i + 1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
    
plt.show() 
```

So <b>in closing, dilation is applied and then followed by erosion.</b>

There are different types of morphological operations, they can be applied using morphologyEx. So, for example, some of them are used, so the mean morphological operations other than opening and closing is morphological gradient.

## Morphological Gradient

```python
# morphological gradient
mg = cv2.morphologyEx(mask, cv2.MORPH_GRADIENT, kernel)

# top-hat
th = cv2.morphologyEx(mask, cv2.MORPH_TOPHAT, kernel)

# see how the image looks like after applying mask and dilation
titles = ['images', 'mask', 'dilation', 'erosion', 'opening', 'closing','mg', 'th']
images = [img, mask, dilation, erosion, opening, closing, mg, th]

# range eight because now the array is of eight elements
# displaying with 2 by 4
for i in range(8):
    plt.subplot(2, 4, i + 1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
    
plt.show() 
```

<b>Morphological gradient</b> is the difference between the dilation and erosion of an image.

<b>Top-hat</b> is the difference between the image and the opening of an image.
